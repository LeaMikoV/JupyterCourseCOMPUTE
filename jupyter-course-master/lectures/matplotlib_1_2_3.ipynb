## Plotting 2D Arrays
\begin{equation}
  f(x,y) =
  \begin{cases}
    xy  & \text{if $x^2+y^2\leq10^6$} \\
    0 & \text{otherwise}
  \end{cases}
\end{equation}

```python
radius = 1000
x = np.linspace(-radius,radius,radius*2+1,dtype=int)
y = x[:,np.newaxis] # y is a column vector
z = np.empty((radius*2+1,radius*2+1))
np.add( x**2, y**2, out=z )
z[z<=radius**2] = (x*y*1e-5)[z<=radius**2]
z[z>radius**2] = 0
```

```python
plt.imshow(z,extent=[-radius,radius,-radius,radius],cmap=plt.get_cmap('viridis'))
plt.colorbar(label=r'$z \times 10^{-3}$  /  $\mu$m$^2$')
plt.xlabel('$x$  /  ')
plt.ylabel('$y$  /  ')
```

```python
plt.imshow(z,extent=[-radius,radius,-radius,radius],cmap=plt.get_cmap('viridis'),origin='lower')
plt.colorbar(label=r'$z \times 10^{-3}$  /  $\mu$m$^2$')
plt.xlabel('$x$  /  ')
plt.ylabel('$y$  /  ')
```

```python
def make_df(df,years):
    df = ( df.xs(['T','NHAZ']).loc[['Glass','Plastic','Paper and cardboard','Metal','Wood'],:]
     .drop(labels=['EU28','IS','LI','NO','ME','MK','RS','TR','XK','BA'],level=2)
    )[years].stack().unstack(1).reset_index(level=[0,2]).copy()
    df['year'] = df['year'].astype(str)
    df.columns.set_names('',inplace=True)
    return df
```
